## Merging 
often we want to join multiple dataframes into one because the data are linked together. This cam be acheived using the `pandas` `merge` function that opperates simmilar to how a merge would work in `R` or in `PowerQuery`. First we specify the two dataframes that we want to merge, the type of merge we want (e.g., left, inner or outer etc),  what variabs are bing matched.

```python
## exersise 1
# merge the child identifies and the child characteristics data together 


# both columns have the LAchildID col;umn as a unique identifier 
child_charateristics = pd.merge(child_identifies, child_charateristics, how="left", on="LAchildID")

child_charateristics
```

as there are two different unamed columns it keeps both, and gives each a suffect sa x or y for lest and right. we can use the suffex argument to specify a name to add to theses columns on eaither dataset. for eaxmpel if we specified `suffix=("_id", "_har")` the the two unamed columns would have thoes appended rather than the `_x` and `_y`. If we wanto to merge based on multiple columns then we just include a list ibn the keyword argument how `on=[var1, var2]`. in this case we can merhe using both matching columns:

```python
child_charateristics = pd.merge(child_identifies, child_charateristics, how="left", on=["LAchildID", "Unnamed: 0"])
child_charateristics
```

## Slicing 
This helps pandas stand out quite a bit, where we take out a selection from the dataframe. This can be don eusing an index locatuion, however it can also be done using logical conditions. 

Logical contions are `+` plus, `-` minus, `<` less than, `>` greater than, `==` equals (boolan chech not assignemty), `~` not, `!=` `&` and, `|` or, `.isna()` is it an NA `.notna()` is it not an Na, `.isin()` ...? , `.str.contains()` does a string contain.



```python
# make a slice of the child_identifiers table and take only children over the mean age using .mean()
 
# we can do this directly in the subsetting 
over_mean = child_identifies[child_identifies["Age"] > child_identifies["Age"].mean()]
over_mean 

# or using a presave condition to subset the data 
over_mean = child_identifies["Age"] > child_identifies["Age"].mean()
child_identifies[over_mean]
```
