Here is a function that invludes a loop and some if / else statements

```python
child_data = {
  'child_id': [3457, 2341, 1111, 4567], 
  'legal_status': ['no', 'yes', 'no', 'yes'],
}


for key_names in child_data:
  if key_names == 'child_id':
    for child_id in child_data[key_names]:
      print('checking', child_id)
      if child_id > 3000:
        print('child_id is greater than 3000')
      else:
        print('child_id is less than 3000')
  elif key_names == 'legal_status':
    legal_statuses = child_data[key_names]
    for legal_status in legal_statuses:
      print(legal_status, len(legal_status))
```
