The central Limit Theorum is a fundimental theorusm in statistiscs. It states that if we take a large number of sampels from a population with finite mean $\mu$ and variance $\sigma^2$, then the distribution of sample means will follow a normal ditribution $N(\mu, \sigma^2)$, regardless of the original distribution. Morover, the mean of saple mean will be equal to the population mean and the standard deviation of the sampel mean - standard error - willdecrease as the sample incraeses.

> 'The Central Limit Theorem explains, at least in part, the prevalence of normal distributions in the natural world. Most characteristics of animals and other life forms are affected by a large number of genetic and environmental factors whose effect is additive. The characteristics we measure are the sum of a large number of small effects, so their distribution tends to be normal.` (Allen Downey)



```python
import numpy as np 
import matplotlib.pyplot as plt 

num = [1, 10, 50, 1000]
means = []

# generating random

for i in num: 
    np.random.seed(1)
    x = [np.mean(
        np.random.randint(
            -100, 100, i)) for _i in range(1000)]
    means.append(x)

k = 0 

fig, ax = plt.subplots(2, 2, figsize =(8, 8))

for i in range(0, 2):
    for j in range(0, 2):
        ax[i, j].hist(means[k], 10, density = True)
        ax[i, j].set_title(label = num[k])
        k += 1

plt.show()

 
```
